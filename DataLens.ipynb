# DataLens: Modular Dataset Explorer

Scans `data`, recursively extracts archives, analyses supported files, and renders representative renderable files.

## Load file-type plugins

```python
import pandas as pd
from IPython.display import display

from datalens.plugin_registry import load_plugins

plugins = load_plugins()

plugin_df = (
    pd.DataFrame(
        {
            "Plugin": [p.name for p in plugins],
            "File Types": [", ".join(sorted(p.suffixes)) for p in plugins],
            "Renderable": ["✅" if p.renderable else "" for p in plugins],
        }
    )
    .sort_values("Plugin")
    .reset_index(drop=True)
)

display(plugin_df)
```

## Set data folder

```python
from pathlib import Path
from IPython.display import display

DATA_DIR = Path('data')
```

## Recursively extract archives

```python
from datalens.archives import extract_archives_recursive

extracted = extract_archives_recursive(DATA_DIR, overwrite=False)
print(f'Extracted/decompressed targets: {len(extracted)}')
```

## Inventory

```python
from datalens.inventory import scan_data_folder, file_type_summary, print_inventory_summary

inventory = scan_data_folder(DATA_DIR)
print_inventory_summary(inventory)
display(inventory.head(10))
display(file_type_summary(inventory))
```

## Analyse supported files

```python
from datalens.plugin_registry import analyse_inventory

results = analyse_inventory(inventory, plugins, max_files_per_type=5)
print(f'Analysis results: {len(results)}')
```

```python
for result in results:
    print('=' * 100)
    print(f"{result.get('plugin')} :: {result.get('path')}")
    for key, value in result.items():
        if key == 'preview':
            print('\npreview:')
            print(value)
```

## Render files

Render files such as shape files `.shp`

```python
from datalens.plugin_registry import render_inventory

render_inventory(inventory, plugins, max_files_per_type=5)
```
